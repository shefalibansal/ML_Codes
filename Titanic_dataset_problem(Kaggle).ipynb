```python
import re
def get_title(name):
    title_search = re.search(' ([A-Za-z]+)\.', name)
 
    if title_search:
        return title_search.group(1)
    return ""

analyse_train['Title'] = analyse_train['Name'].apply(get_title)
merge_data['Title'] = merge_data['Name'].apply(get_title)
print(pd.crosstab(analyse_train['Title'], analyse_train['Sex']))
```

```python
print (analyse_train[['Title', 'Survived']].groupby(['Title'], as_index=False).mean())
```

```python
merge_data['Title'] = merge_data['Title'].replace(['Lady', 'Countess','Capt', 'Col','Don', 'Dr', 'Major', 'Rev', 'Sir', 'Jonkheer', 'Dona'], 'Rare')

merge_data['Title'] = merge_data['Title'].replace('Mlle', 'Miss')
merge_data['Title'] = merge_data['Title'].replace('Ms', 'Miss')
merge_data['Title'] = merge_data['Title'].replace('Mme', 'Mrs')
```

```python
# Mapping titles
title_mapping = {"Mr": 1, "Miss": 2, "Mrs": 3, "Master": 4, "Rare": 5}
merge_data['Title'] = merge_data['Title'].map(title_mapping)
# merge_data['Title'] = merge_data['Title'].fillna(1)
merge_data.drop(['Name'],axis=1,inplace=True)
```
